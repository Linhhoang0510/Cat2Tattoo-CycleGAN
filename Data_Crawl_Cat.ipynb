```python
# Define a function to save images to a specific folder
def save_image_to_folder(image_url, folder, image_number):
  response = requests.get(image_url, stream=True)

  if response.status_code == 200:
    
    # Set up the image file path
    if not os.path.exists(folder):
      os.makedirs(folder)

    # Save the image
    file_path = os.path.join(folder, f'cat_{image_number}.jpg')
    with open(file_path, 'wb') as f:
      for chunk in response.iter_content(1024):
        f.write(chunk)

    print(f"Downloaded image {image_number}")

  else:
    print(f"Failed to download image {image_number}, status code {response.status_code}")
```

```python
import requests
import time
import json
import os

time.sleep(0.1)

# Insert API key here
access_key = ''

# Spcify collection id to download images from
collection_id = ['3816160', '10029777']

# Define a function to download images from collections
def download_images_from_collection(collection_id, folder):
    image_count = 0
    page = 1

    while True:
        # The base URL for the Unsplash API for a specific collection
        URL = f"https://api.unsplash.com/collections/{collection_id}/photos"

        # Parameters for the API request
        params = {
            'Accept-Version': 'v1',
            'client_id': access_key,
            'per_page': 50,
            'page': page
        }

        # Get the response
        response = requests.get(URL, params=params)

        if response.status_code == 200:
            photos = response.json()

            if not photos:
                break  # No more photos in the collection, exit the loop

            for photo in photos:
                image_count += 1
                image_url = photo['urls']['regular']
                
                save_image_to_folder(image_url, folder, image_count) # USing function defined above
            
            page += 1  # Go to the next page
            
            time.sleep(0.1)  # Sleep to obey rate limits
        
        else:
            print(f'Failed to fetch collection {collection_id}, status code {response.status_code}')
            break
```

```python
# Iterate through collection id
for id in collection_id:
  folder_name = f'collection_{id}_image' # Specify folder name
  download_images_from_collection(id, folder_name)
```
